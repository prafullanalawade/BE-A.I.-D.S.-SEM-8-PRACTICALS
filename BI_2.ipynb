```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Satyam\Downloads\sales_data_sample.csv", encoding= 'latin1')  # Replace with your dataset path
data
```

```python
print("Data Extracted ✅")
```

```python
# Step 2: Transform
data.columns = data.columns.str.strip()
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
data['Month'] = data['ORDERDATE'].dt.to_period('M')
monthly_sales = data.groupby('Month')['SALES'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)
```

```python
# Load into SQLite
conn = sqlite3.connect('sales_db.sqlite')
monthly_sales.to_sql('monthly_sales', conn, if_exists='replace', index=False)
```

```python
# Step 4: Visualization
# Visualize
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales['Month'], monthly_sales['SALES'], marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```
